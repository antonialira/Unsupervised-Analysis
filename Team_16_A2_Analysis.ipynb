<h4>d) User-Defined Functions</h4><br>
Run the following code to load the user-defined functions used throughout this Notebook.

```python
# user-defined functions

########################################
# scree_plot --> Source: Chase Kusterer
########################################
def scree_plot(pca_object, figure_size,  export = False):
    """
    Visualizes a scree plot from a pca object.
    
    PARAMETERS
    ----------
    pca_object | A fitted pca object
    figure_size| Set the size of the figure (width, height)
    export     | Set to True if you would like to save the scree plot to the
               | current working directory (default: False)
    """
    # building a scree plot

    # setting plot size
    fig, ax = plt.subplots(figsize=figure_size)
    features = range(pca_object.n_components_)


    # developing a scree plot
    plt.plot(features,
             pca_object.explained_variance_ratio_,
             linewidth = 2,
             marker = 'o',
             markersize = 10,
             markeredgecolor = 'black',
             markerfacecolor = 'grey')


    # setting more plot options
    plt.title('Scree Plot')
    plt.xlabel('PCA feature')
    plt.ylabel('Explained Variance')
    plt.xticks(features)

    if export == True:
    
        # exporting the plot
        plt.savefig('./__analysis_images/top_customers_correlation_scree_plot.png')
        
    # displaying the plot
    plt.show()


########################################
# unsupervised_scaler --> Source: Chase Kusterer
########################################
def unsupervised_scaler(df):
    """
    Standardizes a dataset (mean = 0, variance = 1). Returns a new DataFrame.
    Requires sklearn.preprocessing.StandardScaler()
    
    PARAMETERS
    ----------
    df     | DataFrame to be used for scaling
    """

    # INSTANTIATING a StandardScaler() object
    scaler = StandardScaler()


    # FITTING the scaler with the data
    scaler.fit(df)


    # TRANSFORMING our data after fit
    x_scaled = scaler.transform(df)

    
    # converting scaled data into a DataFrame
    new_df = pd.DataFrame(x_scaled)


    # reattaching column names
    new_df.columns = df.columns
    
    return new_df

########################################
# explained variance
########################################

def explained_variance(pca):
    
    # Counter
    counter = 0

    # Lists for variance
    variance_lst = []
    component_number = []
    cumulative = []
    delta = []

    # looping over each principal component
    for variance in pca.explained_variance_ratio_:

        variance_lst.append(variance)

        # conditional to fill each list
        if counter == 0:

            component_number.append(0)
            cumulative.append(variance)        
            delta.append(0) 

        else:

            component_number.append(counter)
            cumulative.append(cumulative[counter-1] + variance)
            delta.append(variance_lst[counter-1] - variance_lst[counter])

        # increase the counter by 1
        counter += 1    

    # round each list
    variance_lst = [round(num, 4) for num in variance_lst]
    cumulative   = [round(num, 4) for num in cumulative]
    delta        = [round(num, 4) for num in delta]

    # dictionary of lists 
    variance_data = {'PC': component_number, 
                     'Variance': variance_lst, 
                     'Cumulative': cumulative, 
                     'Delta': delta} 

    # creating a dataframe with the dictionary
    variance_df = pd.DataFrame(variance_data)

    # checking the DF
    return variance_df
```
